```python
# Adding a new column to calculate the difference between price of observation and mean price of data set

def diff(x):
    mean=RE.PRICE.mean()
    diff=x-mean
    return diff 

RE['meandiff']=RE.PRICE.apply(diff)
RE.head()
```

### Step 6: Running Decision Tree

```python
TreeColumns = ['GROSSSIZE', 'YEARBUILT', 'latitude', 'longitude']

X = RE[TreeColumns]
y = RE.PRICE
```

```python
# Spliting into train/test

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Printing out each of the arrays

print X_train
print y_train
print X_test
print y_test
```

```python
# Instantiate estimator and fitting with training set

from sklearn.tree import DecisionTreeRegressor
treereg = DecisionTreeRegressor(random_state=1)
treereg.fit(X_train, y_train)
```

```python
# Generating Predictions
preds = treereg.predict(X_test)

# Printing predictions and actual values
print preds
print y_test
```

```python
# Wow - very large Mean Squared Error, however should not be surprising given low correlation demonstrated with 
# Linear Regression

from sklearn import metrics
import numpy as np
np.sqrt(metrics.mean_squared_error(y_test, preds))
```

```python
# Using cross-validation to find best max_depth

from sklearn.cross_validation import cross_val_score
```

```python
% matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

all_scores = []
best_score = -1
best_depth = 0
for i in range(1, 6):
    treereg = DecisionTreeRegressor(max_depth=i, random_state=1)
    scores = cross_val_score(treereg, X, y, cv=3, scoring='mean_squared_error')
    current_score = np.mean(np.sqrt(-scores))
    # If the score mean is better than the current best, or best is the default (-1), then update!
    if current_score < best_score or best_score == -1:
        best_score = current_score
        best_depth = i
    # store to plot anyway!
    all_scores.append(current_score)
    
print "Best score: %s" % best_score
print "Best depth: %s" % best_depth

# now actually fit the model
treereg = DecisionTreeRegressor(max_depth=best_depth, random_state=1)
treereg.fit(X, y)

plt.figure()
plt.plot(range(1, 6), all_scores)
plt.xlabel('x=max tree depth')
```

```python
from sklearn.tree import export_graphviz
with open("RE.dot", 'wb') as f:
    f = export_graphviz(treereg, out_file=f, feature_names=TreeColumns)
```
