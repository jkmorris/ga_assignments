```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from __future__ import division

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
#Read Me: 

# My project attempts to build a model that allows the New York State Department of Taxation and Finance 
# to better understand the factors that influence state sales tax. My hypothesis is that the state treasurer can better
# budget and match planned expenses with future sales tax revenue by using a predictive model based on leading 
# economic indicators. 

# While the number of NYC Tourists, estimated taxes generated by tourists and total wages of NY-state employees 
# use actual numbers, the variables Unemployement and the Dow Jones index (a proxy for the stock market) are based on 
# an index where 2nd Quarter 2003 is set to 100 as I believe it is not the overall level of Unemployment and the 
# stock market that drives consumer spending, but the relative quarter-to-quarter change in these indices. 

# Note: I have set the Unemployment Index to increase as the Unemployment rate goes down (and vice versa). 
```

```python
tax=pd.read_csv("NYTax.csv")
```

```python
tax.describe()
```

```python
# Scatter Matrix

cols = ['SalesTax', 'TouristTax', 'NYWage', 'DJIndexed', 'UnemploymentIndexed',]
pd.scatter_matrix(tax[cols], figsize=(12, 12))

# Generally appears to have positive correlation across variables 
```

```python
import statsmodels.formula.api as smf

NYTax = pd.DataFrame(tax)

fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(NYTax.SalesTax, NYTax.TouristTax, 'g.')

model = smf.ols(formula='SalesTax ~ TouristTax + NYWage + DJIndexed + UnemploymentIndexed', data=NYTax)
results = model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())
print

# Note: While the R-Squared is large, the p-value for NYWage variable is greater than 5%.

# Note the warning for possibility of Multicollinearity 
```

```python
# Rerunning model by excluding NYWage variable

refined_model = smf.ols(formula='SalesTax ~ TouristTax + DJIndexed + UnemploymentIndexed', data=NYTax)
results = refined_model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())
print
```
